# Visualization of progression Graph (in R)
(https://sites.google.com/site/centretranslationinnovation/tpr-db/visualization-in-r)
- read the source file (progGra.R, or app.R)
    - two functions:
    - ReadData(<filename>) : reads data to   
    - ProgGraph() : plot the progression graph
- read data from a session
- vizualize the data with a progression graph
- modify settings


```r
# read the R core  with the functions
source("/data/critt/shiny-server/PG/Annotate/app.R")

# this reads a default session
```

```r
# read the session that we want to vizualize
ReadData("/data/critt/tprdb/TPRDB/BML12/Tables/P06_T5")
```

```r
# plot a view of the entire session
ProgGraph()
```

```r
# zoom into the session, select a time window 
ProgGraph(X1=30000, X2=80000)
```

```r
# zoom into the session, select a time window (X) and a text window (Y) 
ProgGraph(X1 = 45000, X2 =55000, Y1=24, Y2=35)
```

```r
# visualise production unit (pu=1)
ProgGraph(X1 = 45000, X2 =55000, Y1=24, Y2=35, pu=1, fu=1)
```

```r
# zoom further into the segment, visualise production units and fixation units 
ProgGraph(X1 = 45000, X2 =55000, Y1=24, Y2=35, pu=1, fu=1, au=1, auLines = 1, seg = 0, label = 0)
```
